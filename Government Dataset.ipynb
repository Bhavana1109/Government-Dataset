### ANALYSIS ON GOVERNMENT DATA SET

### Analysis of Consumer Price Index (CPI)

### INTRODUCTION:
This dataset contains the All India Consumer Price Index (CPI) data as of February 2019. It provides monthly price index values for various categories of goods and services across different sectors (Rural, Urban, and Combined). The dataset spans multiple years and includes different consumer expenditure categories such as:
Food Items: Cereals, Meat, Fish, Eggs, Milk, Oils, Fruits, Vegetables, Sugar, and Spices.
Non-Food Items: Clothing, Footwear, Housing, Fuel, Household Goods, Health, Transport, Education, and Miscellaneous expenses.
General Index: A summary index reflecting overall price levels.
Each row represents CPI values for a given year, month, and sector (Rural, Urban, or Combined). The dataset is structured with 30 columns and 222 records, though some values (e.g., Housing) have missing data.

```python
import pandas as pd
df = pd.read_csv("All_India_Index_feb2019_0.csv")
df
```

This dataset has 222 rows and 30 columns.

In this dataset missing values present in Housing column.

### Column Descriptions:
1. Sector: Type of region (Rural, Urban, or Rural+Urban).
2. Year: The year of data recording.
3. Month: The month of data recording.
4. Cereals and products: Index for cereal-based food items.
5. Meat and fish: Index for meat and fish products.
6. Egg: Index for eggs.
7. Milk and products: Index for milk and dairy products.
8. Oils and fats: Index for edible oils and fats.
9. Fruits: Index for various fruits.
10. Vegetables: Index for various vegetables.
11. Pulses and products: Index for pulses (lentils, beans, etc.).
12. Sugar and Confectionery: Index for sugar and related products.
13. Spices: Index for commonly used spices.
14. Non-alcoholic beverages: Index for soft drinks, tea, coffee, etc.
15. Prepared meals, snacks, sweets etc.: Index for packaged and prepared food items.
16. Food and beverages: Combined index for all food and beverage categories.
17. Pan, tobacco and intoxicants: Index for tobacco, betel leaves, and similar products.
18. Clothing: Index for clothing items.
19. Footwear: Index for footwear.
20. Clothing and footwear: Combined index for clothing and footwear.
21. Housing: Index for housing-related expenses (only available for Urban & Rural+Urban sectors).
22. Fuel and light: Index for fuel (LPG, electricity, kerosene, etc.).
23. Household goods and services: Index for furniture, appliances, and home services.
24. Health: Index for medical expenses.
25. Transport and communication: Index for public and private transportation, and communication services.
26. Recreation and amusement: Index for entertainment expenses.
27. Education: Index for education-related expenses.
28. Personal care and effects: Index for personal care products (cosmetics, hygiene products, etc.).
29. Miscellaneous: Index for other categories not explicitly mentioned.
30. General index: Overall price index covering all categories.

```python
df["Month"] = df["Month"].str.strip().replace("Marcrh", "March")
df["Month"] = df["Month"].apply(lambda x: x.capitalize() if isinstance(x, str) else x)
```

```python
df["Date"] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Month"], format="%Y-%B", errors='coerce')
```

```python
df = df.dropna(subset=["Date"]).copy()
```

```python
df.drop(columns=["Year", "Month"], inplace=True)
```

Removes 'Year' and 'Month' columns after creating a combined 'Date' column.

```python
df.loc[:, "Housing"] = df["Housing"].fillna(df["Housing"].median())
```

Missing values in the 'Housing' column are filled with the median value to maintain data consistency.

```python
df = df.drop_duplicates().copy()
```

### Removing Duplicates
Ensures data integrity by dropping duplicate rows.
